```python
#Chap02_Cm01

import numpy as np

x = np.random.uniform(0.0, 1.0, 10000)
y = np.random.uniform(0.0, 1.0, 10000)
```

```python
#Chap02_Cm02

import math
import matplotlib
import matplotlib.pyplot as plt

inside = 0

inside_circle_x = []
inside_circle_y = []

outside_circle_x = []
outside_circle_y = []

for i in range(0, 10000):
  d = math.hypot(x[i],y[i])
  if d < 1:
    inside += 1
    inside_circle_x.append(x[i])
    inside_circle_y.append(y[i])
  else:
    outside_circle_x.append(x[i])
    outside_circle_y.append(y[i])

print("円内部の点の数:", inside)

Circle = matplotlib.patches.Circle(xy=(0,0), radius=1, fill = None)

fig, ax = plt.subplots(figsize=(6, 6))
ax.add_patch(Circle)
ax.autoscale()
ax.grid()

plt.scatter(inside_circle_x,inside_circle_y)
plt.scatter(outside_circle_x,outside_circle_y)
plt.show()
```

```python
#Chap02_Cm03

print("円周率:", 4*inside/10000)
```
